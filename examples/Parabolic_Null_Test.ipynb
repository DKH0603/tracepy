```python
import numpy as np
import matplotlib.pyplot as plt

import tracepy as tp
```

```python
mirror = {
    'action': 'reflection',
    'P': 1.5,
    'kappa': 0.,
    'c': -.5,
    'Diam': 2.2,
}

stop = {
    'action': 'stop',
    'P': .5,
    'Diam': .2
}

geo = [mirror, stop]
ray_group = tp.ray_plane(geo, [0., 0., -1.5], 1.1, d=[0., 0., 1.], nrays=50)
```

```python
tp.plotyz(geo, ray_group)
```

```python
tp.spotdiagram(geo, ray_group)
```

```python
tp.rayaberration(geo, ray_group)
```
