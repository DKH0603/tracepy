```python
import numpy as np
import matplotlib.pyplot as plt

import codevp as vp
```

```python
lens = {
    'name': 'lens',
    'action': 'refraction',
    'P': 2.,
    'kappa': -.004,
    'c': -.5,
    'Diam': 2.2
}

back_lens = {
    'name': 'back_lens',
    'action': 'refraction',
    'P': 1.,
    'c':.5,
    'kappa': -.004,
    'N': 1.5,
    'Diam': 2.2
}

stop = {
    'name': 'stop',
    'action': 'stop',
    'P': 3.,
    'Diam': 2.2
}

geo_1 = [back_lens, lens, stop]
ray_group = vp.ray_plane(geo_1, [0., 0., 0.], 1.1, d=[0.,0.,1.], nrays=100)
```

```python
rayparams = {'color': 'red',
             'alpha': 0.3}

gplt = vp.geoplot(geo_1, ray_group, pltparams=rayparams)
oplt = vp.optplot(geo_1, ray_group)
```

```python
gplt.plotyz()
```

```python
oplt.spotdiagram()
```

```python
vlens = {
    'name': 'stop',
    'vary': ['P']
}

vlens2 = {
    'name': 'back_lens',
    'vary': ['c', 'kappa', 'N']
}

vlens3 = {
    'name': 'lens',
    'vary': ['c', 'kappa']
}

vary_list = [vlens, vlens2, vlens3]
```

```python
geo_opt = vp.optimize(geo_1, vary_list, typeof='least_squares')
```

```python
ray_group_opt = vp.ray_plane(geo_opt, [0., 0., 0.], 1.1, d=[0.,0.,1.], nrays=100)
gplt_opt = vp.geoplot(geo_opt, ray_group_opt, pltparams=rayparams)
oplt_opt = vp.optplot(geo_opt, ray_group_opt)
```

```python
gplt_opt.plotyz()
```

```python
oplt_opt.spotdiagram()
```
