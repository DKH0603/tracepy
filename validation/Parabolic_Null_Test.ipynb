```python
from numpy import pi

import numpy as np
import matplotlib.pyplot as plt

import codevp as vp
from codevp import ray_plane, save_optics
```

```python
#Parabolic null test
mirror = {
    'shape': 'conic',
    'inter': 'reflection',
    'D': np.array([0.,0.,0.]),
    'P': np.array([0.,0.,1.5]),
    'kappa': 0.,
    'c': -.5,
    'r': 1.1
}

stop = {
    'shape': 'plane',
    'inter': 'stop',
    'P': np.array([0., 0., .5]),
    'D': np.array([0., 0., 0.]),
    'r': .1
}

geo = [mirror, stop]
ray_group = ray_plane(geo, [0., 0., -1.5], 1.1, d=[0.,0.,1.], nrays=50)
```

```python
rayparams = {'color': 'red',
             'alpha': 0.3}

gplt = vp.geoplot(geo, ray_group, pltparams=rayparams)
oplt = vp.optplot(geo, ray_group)
```

```python
plt.figure(dpi=200)
gplt.plot2d()
```

```python
plt.figure(dpi=150)
oplt.spotdiagram()
plt.show()
```
