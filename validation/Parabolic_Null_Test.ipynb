```python
from numpy import pi

import numpy as np
import matplotlib.pyplot as plt

import codevp as vp
```

```python
#Parabolic null test
mirror = {
    'shape': 'conic',
    'inter': 'reflection',
    'D': np.array([0.,0.,0.]),
    'P': np.array([0.,0.,1.5]),
    'kappa': 0.,
    'c': -.5,
    'Diam': 2.2,
}

stop = {
    'shape': 'plane',
    'inter': 'stop',
    'P': np.array([0., 0., .5]),
    'D': np.array([0., 0., 0.]),
    'Diam': .2
}

geo = [mirror, stop]
ray_group = vp.ray_plane(geo, [0., 0., -1.5], 1.1, d=[0., 0., 1.], nrays=50)
```

```python
rayparams = {'color': 'red',
             'alpha': 0.3}

gplt = vp.geoplot(geo, ray_group, pltparams=rayparams)
oplt = vp.optplot(geo, ray_group)
```

```python
gplt.plotyz()
```

```python
oplt.spotdiagram()
plt.show()
```
