```python
import numpy as np
import matplotlib.pyplot as plt

import raypy as ray
```

```python
#Parabolic null test
mirror = {
    'action': 'reflection',
    'P': 1.5,
    'kappa': 0.,
    'c': -.5,
    'Diam': 2.2,
}

stop = {
    'action': 'stop',
    'P': .5,
    'Diam': .2
}

geo = [mirror, stop]
ray_group = ray.ray_plane(geo, [0., 0., -1.5], 1.1, d=[0., 0., 1.], nrays=50)
```

```python
rayparams = {'color': 'red',
             'alpha': 0.3}

gplt = ray.geoplot(geo, ray_group, pltparams=rayparams)
oplt = ray.optplot(geo, ray_group)
```

```python
gplt.plotyz()
```

```python
oplt.spotdiagram()
plt.show()
```
