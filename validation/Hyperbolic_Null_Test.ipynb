```python
from numpy import pi

import numpy as np
import matplotlib.pyplot as plt

import codevp as vp
from codevp import ray_plane, save_optics
```

```python
#Hyperbolic null test
rad = 1.1

lens = {
    'shape': 'conic',
    'inter': 'refraction',
    'D': np.array([0.,0.,0.]),
    'P': np.array([0.,0.,2.]),
    'kappa': -2.25+1,
    'c': -2,
    'r': rad
}

back_lens = {
    'shape': 'plane',
    'inter': 'refraction',
    'D': np.array([0.,0.,0.]),
    'P': np.array([0.,0.,1]),
    'N': 1.5,
    'r': rad
}

stop = {
    'shape': 'plane',
    'inter': 'stop',
    'P': np.array([0., 0., 3.]),
    'D': np.array([0., 0., 0.]),
    'r': rad
}

geo = [back_lens, lens, stop]
ray_group = ray_plane(geo, [0., 0., 0.], rad, d=[0.,0.,1.], nrays=100)
```

```python
rayparams = {'color': 'red',
             'alpha': 0.3}

gplt = vp.geoplot(geo, ray_group, pltparams=rayparams)
oplt = vp.optplot(geo, ray_group)
```

```python
plt.figure(dpi=200)
gplt.plot2d()
```

```python
plt.figure(dpi=150)
oplt.spotdiagram()
```
