```python
from numpy import pi

import numpy as np
import matplotlib.pyplot as plt

import codevp as vp
```

```python
#Hyperbolic null test
Diam = 2.2

lens = {
    'shape': 'conic',
    'inter': 'refraction',
    'D': np.array([0.,0.,0.]),
    'P': np.array([0.,0.,2.]),
    'kappa': -2.25+1,
    'c': -2,
    'Diam': Diam
}

back_lens = {
    'shape': 'plane',
    'inter': 'refraction',
    'D': np.array([0.,0.,0.]),
    'P': np.array([0.,0.,1.]),
    'N': 1.5,
    'Diam': Diam
}

stop = {
    'shape': 'plane',
    'inter': 'stop',
    'P': np.array([0., 0., 3]),
    'D': np.array([0., 0., 0.]),
    'Diam': Diam
}

geo = [back_lens, lens, stop]
ray_group = vp.ray_plane(geo, [0., 0., 0.], Diam/2., d=[0.,0.,1.], nrays=100)
```

```python
rayparams = {'color': 'red',
             'alpha': 0.3}

gplt = vp.geoplot(geo, ray_group, pltparams=rayparams)
oplt = vp.optplot(geo, ray_group)
```
