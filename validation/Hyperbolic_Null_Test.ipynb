```python
import numpy as np
import matplotlib.pyplot as plt

import codevp as vp
```

```python
#Hyperbolic null test
lens = {
    'action': 'refraction',
    'P': 2.,
    'kappa': -1.25,
    'c': -2,
    'Diam': 2.2
}

back_lens = {
    'action': 'refraction',
    'P': 1.,
    'N': 1.5,
    'Diam': 2.2
}

stop = {
    'action': 'stop',
    'P': 3.,
    'Diam': 2.2
}

geo = [back_lens, lens, stop]
ray_group = vp.ray_plane(geo, [0., 0., 0.], 1.1, d=[0.,0.,1.], nrays=100)
```

```python
rayparams = {'color': 'red',
             'alpha': 0.3}

gplt = vp.geoplot(geo, ray_group, pltparams=rayparams)
oplt = vp.optplot(geo, ray_group)
```
